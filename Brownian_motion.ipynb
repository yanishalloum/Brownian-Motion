```python
import numpy as np
import matplotlib.pyplot as plt

# T = borne supérieure de l'espace temporelle [0, T]
# N = nombre de points de discrétisation de l'espace temporelle

def bm(T, N=1000):
    # Incrément infitésimal temporel
    dt = T / N

    # Incrément infinitésimal de processus gaussiens indépendants
    dW = np.sqrt(dt) * np.random.randn(N)

    # Somme cumulative des incréments gaussiens -> trajectoire
    W = np.cumsum(dW)

    # Initialisation à W(0) = 0
    W = np.concatenate(([0], W))

    # Axe temporel
    t = np.linspace(0, T, N+1)

    # Plot de la trajectoire
    plt.figure(figsize=(10, 6))
    plt.plot(t, W)
    plt.title('Trajectoire du mouvement brownien standard')
    plt.xlabel('t')
    plt.ylabel('W(t)')
    plt.grid(True)
    plt.show()

# Illustration
bm(1)
```

La fonction tracée est bien presque sûrement continue partout et d'espérance nulle (centrée autour de 0). Le mouvement ne semble pas être à variation infinie car la courbe ne change pas de signe un grand nombre de fois au voisinage de 0. Cela est sûrement dû au fait que la discrétisation est trop faible.

```python
import numpy as np
import matplotlib.pyplot as plt

# T: borne supérieure de l'espace temporel
# mu: drift/dérive du mouvement brownien (MB)
# sigma^2 : variance du MB
# N : nombre de points de discrétisation
# x : initialisation du MB (X[0] = x)
def bm_drifted(T, mu, sigma, x, N=1000):
    # Déplacement temporel infinitésimal
    dt = T / N

    # Déplacement infinitésimal de processus gaussiens indépendants
    dW = np.sqrt(dt) * np.random.randn(N)

    # Somme cumulative des incréments gaussiens -> trajectoire
    W = np.cumsum(dW)

    # Trajectoire de X(t) avec dérive, variance et initialisation à x
    t = np.linspace(0, T, N+1)
    X = x + mu * t[1:] + sigma * W

    # Ajout de X(0) = x au début
    X = np.concatenate(([x], X))

    # Tracé de la trajectoire
    plt.figure(figsize=(10, 6))
    plt.plot(t, X)
    plt.title('Trajectoire du mouvement brownien avec dérive et variance')
    plt.xlabel('Temps t')
    plt.ylabel('X(t)')
    plt.grid(True)
    plt.show()

# Illustration
bm_drifted(T=1, mu=2, sigma=20, x=1)
```

On constate effectivement que les 3 valeurs obtenues sont similaires, vérifiant numériquement le principe de réflexion.

```python
import numpy as np


a = 3
b = 5
n = 5000
dt = 0.01

# Simuler des trajectoires de mouvement brownien
L = [] # liste pour stocker les temps et les valeurs du mouvement brownien où -a ou b sont atteints

# boucle for pour créer n marches aléatoires
for k in range(n):
    # initialisation de Bt et t à 0
    Bt = 0
    t = 0
    # tant que le mouvement brownien n'atteint pas -a ou b, on continue le déplacement
    while True:
        Bt += np.random.normal(0, np.sqrt(dt))
        t += dt
        if Bt <= -a or Bt >= b:
            break
    if Bt >= b: # on stocke uniquement les valeurs de t et Bt lorsque Bt >= b dans la liste L
      L.append((t,Bt))

# Calcul de la probabilité d'atteindre b avant -a
taille = len(L) # on compte le nombre d'élements lorsque Bt >= b
P_Tb = taille / n # on divise par le nombre de simulation pour avoir le ratio expérimental
P_Tb_théorique = a / (a + b) # P( T = Tb ) = a / (a+b)

# Afficher les résultats
print(f"P(T = T_b) expérimental = {P_Tb}")
print(f" P(T = T_b) théorique = {P_Tb_théorique}")
```

Par la méthode de Monte Carlo, on observe que P(T = Tb) expérimental converge vers la valeur théorique de P( T = Tb ) à savoir a/(a+b).

```python
import numpy as np
import matplotlib.pyplot as plt

def interpolate_and_normalize_trajectory(x, N):
    interpolated = []
    for t in np.linspace(0, 1, N):
        nt = t * N
        k = int(np.floor(nt))
        fraction = nt - k
        if k == 0:
            interpolated.append(0)
        elif k < N:
            value = (1 / np.sqrt(N)) * (np.sum(x[:k]) + fraction * x[k])
            interpolated.append(value)
        else:
            value = (1 / np.sqrt(N)) * np.sum(x)
            interpolated.append(value)
    return np.array(interpolated)

N = 1000  # Nombre de pas de la marche aléatoire

# Variables aléatoires i.i.d. centrées et réduites
X = np.random.normal(0, 1, N)

# Interpoler et normaliser la trajectoire
trajectory = interpolate_and_normalize_trajectory(X, N)

# Tracer la trajectoire
plt.figure(figsize=(12, 8))
plt.plot(np.linspace(0, 1, N), trajectory, label='Trajectoire interpolée et normalisée')

plt.title('Trajectoire normalisée et interpolée de la marche aléatoire pour N = {}'.format(N))
plt.xlabel('Temps (t)')
plt.ylabel('Xn(t)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def interpolate_and_normalize_trajectory(x, N):
    interpolated = []
    for t in np.linspace(0, 1, N):
        nt = t * N
        k = int(np.floor(nt))
        fraction = nt - k
        if k == 0:
            interpolated.append(0)
        elif k < N:
            value = (1 / np.sqrt(N)) * (np.sum(x[:k]) + fraction * x[k])
            interpolated.append(value)
        else:
            value = (1 / np.sqrt(N)) * np.sum(x)
            interpolated.append(value)
    return np.array(interpolated)

N = 10000  # Nombre de pas de la marche aléatoire (suffisamment grand pour illustrer la convergence)
num_trajectories = 10  # Nombre de trajectoires à tracer

# Génération des variables aléatoires i.i.d. centrées et réduites
X = np.random.normal(0, 1, (num_trajectories, N))

# Interpoler et normaliser les trajectoires
trajectories = []
for i in range(num_trajectories):
    trajectory = interpolate_and_normalize_trajectory(X[i], N)
    trajectories.append(trajectory)

# Convertir en tableau numpy pour faciliter les calculs
trajectories = np.array(trajectories)

# Calculer la moyenne des trajectoires
mean_trajectory = np.mean(trajectories, axis=0)

# Tracer les trajectoires individuelles et la moyenne
plt.figure(figsize=(12, 8))
for i in range(num_trajectories):
    plt.plot(np.linspace(0, 1, N), trajectories[i], alpha=0.5)

plt.plot(np.linspace(0, 1, N), mean_trajectory, label='Moyenne des trajectoires', color='black', linewidth=2)

plt.title('Trajectoires normalisées et interpolées de la marche aléatoire (illustration du Théorème de Donsker)')
plt.xlabel('Temps (t)')
plt.ylabel('Xn(t)')
plt.legend()
plt.grid(True)
plt.show()
```

Plus le nombre de trajectoires augmente, plus la moyenne se stabilise à zéro. Le mouvement brownien est un processus stochastique centré, i.e. ∀t≥0,E[Bt]=0.La figure confirme cette propriété. Cela illustre bien le théorème de Donsker.

```python
import numpy as np
import matplotlib.pyplot as plt

# Fonction pour générer un mouvement brownien standard
def mb(T, N):
    dt = T / N
    dW = np.sqrt(dt) * np.random.randn(N)
    W = np.cumsum(dW)
    W = np.concatenate(([0], W))  # Initialiser à W(0) = 0
    return W

# Fonction pour tracer des trajectoires du mouvement brownien géométrique
def plot_mb_geometric(T, mu, sigma, N=1000, num_trajectories=5):
    t = np.linspace(0, T, N+1)

    plt.figure(figsize=(12, 8))

    for _ in range(num_trajectories):
        Bt = mb(T, N)
        S0 = 1 #np.random.uniform(0, 1)
        St = S0 * np.exp(mu * t - 0.5 * sigma**2 * t + sigma * Bt)
        plt.plot(t, St)

    plt.title('Trajectoires du mouvement brownien géométrique')
    plt.xlabel('Temps (t)')
    plt.ylabel('S_t')
    plt.grid(True)
    plt.show()

# Paramètres
T = 50     # Borne supérieure de l'intervalle de temps
mu = 0.2      # Drift
sigma = 0.2    # Coefficient de diffusion
N = 100000       # Nombre de points de discrétisation
num_trajectories = 5  # Nombre de trajectoires à tracer

# Plot les trajectoires
plot_mb_geometric(T, mu, sigma, N, num_trajectories)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def pontbrownien(n, npas): # npas, nombre de pas et n, nombre de simulations donc de trajectoires
    dt = 1 / npas # calcul du pas en temps
    t = np.linspace(0, 1, npas + 1) # subdivision de l'intervalle [0,1]

    # Simuler n trajectoires de mouvement brownien
    B = np.random.normal(0, np.sqrt(dt), (n, npas)) # mouvement brownien classique
    B = np.cumsum(B, axis=1)                        # somme cumulée pour obtenir les trajectoires
    B = np.hstack((np.zeros((n, 1)), B))            # On ajoute la condition B à t=0 vaut 0 pour chaque trajectoire

    # On crée le pont brownien en fixant B à t=1 , à la valeur 0
    pont = B - t * B[:, -1][:, np.newaxis] # Comme Bt est un mouvement brownien, alors Bt - tB1 est un pont brownien d'après le cours
    # on rajoute [:, np.newaxis] par soucis de dimension

    return t, pont # on retourne la liste des temps et le pont brownien correspondant

# Paramètres de test
npas = 1000
n = 3

t, pont = pontbrownien(n, npas)

# Tracer les trajectoires
for i in range(n):
    plt.plot(t, pont[i])
plt.plot(t,np.zeros(npas+1), color='g')
plt.title('Trajectoires de pont brownien')
plt.xlabel('Temps')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Fonction pour générer un pont brownien sur [0, 1]
def brownian_bridge(T, N):
    dt = T / N
    t = np.linspace(0, T, N+1)
    W = np.random.normal(0, np.sqrt(dt), N)
    W = np.cumsum(W)
    W = np.concatenate(([0], W))
    W = W - t / T * W[-1]  # Adjust to make it a bridge
    return W

# Fonction pour générer une excursion brownienne en utilisant la transformée de Verwaat
def brownian_excursion(T, N):
    W = brownian_bridge(T, N)
    t = np.linspace(0, T, N+1)
    
    # Trouver l'argument du minimum de W
    Tmin = np.argmin(W)
    
    # Générer l'excursion brownienne Z_t
    Z = np.zeros_like(W)
    for i in range(N+1):
        if i <= Tmin:
            Z[i] = W[Tmin + i] - W[Tmin]
        else:
            Z[i] = W[Tmin + i - (N+1)] - W[Tmin]
    
    return t, Z

# Paramètres
T = 1.0        # Borne supérieure de l'intervalle de temps
N = 1000       # Nombre de points de discrétisation
num_trajectories = 5  # Nombre de trajectoires à tracer

# Tracer les trajectoires de l'excursion brownienne
plt.figure(figsize=(12, 8))
for _ in range(num_trajectories):
    t, Z = brownian_excursion(T, N)
    plt.plot(t, Z)

plt.title('Trajectoires de l\'excursion brownienne')
plt.xlabel('Temps (t)')
plt.ylabel('Z_t')
plt.grid(True)
plt.show()
```
