```python
import numpy as np
import matplotlib.pyplot as plt

# T = borne suprieure de l'espace temporelle [0, T]
# N = nombre de points de discrtisation de l'espace temporelle

def bm(T, N=1000):
    # Incrment infitsimal temporel
    dt = T / N

    # Incrment infinitsimal de processus gaussiens indpendants
    dW = np.sqrt(dt) * np.random.randn(N)

    # Somme cumulative des incrments gaussiens -> trajectoire
    W = np.cumsum(dW)

    # Initialisation  W(0) = 0
    W = np.concatenate(([0], W))

    # Axe temporel
    t = np.linspace(0, T, N+1)

    # Plot de la trajectoire
    plt.figure(figsize=(10, 6))
    plt.plot(t, W)
    plt.title('Trajectoire du mouvement brownien standard')
    plt.xlabel('t')
    plt.ylabel('W(t)')
    plt.grid(True)
    plt.show()

# Illustration
bm(1)
```

La fonction trace est bien presque srement continue partout et d'esprance nulle (centre autour de 0). Le mouvement ne semble pas tre  variation infinie car la courbe ne change pas de signe un grand nombre de fois au voisinage de 0. Cela est srement d au fait que la discrtisation est trop faible.

```python
import numpy as np
import matplotlib.pyplot as plt

# T: borne suprieure de l'espace temporel
# mu: drift/drive du mouvement brownien (MB)
# sigma^2 : variance du MB
# N : nombre de points de discrtisation
# x : initialisation du MB (X[0] = x)
def bm_drifted(T, mu, sigma, x, N=1000):
    # Dplacement temporel infinitsimal
    dt = T / N

    # Dplacement infinitsimal de processus gaussiens indpendants
    dW = np.sqrt(dt) * np.random.randn(N)

    # Somme cumulative des incrments gaussiens -> trajectoire
    W = np.cumsum(dW)

    # Trajectoire de X(t) avec drive, variance et initialisation  x
    t = np.linspace(0, T, N+1)
    X = x + mu * t[1:] + sigma * W

    # Ajout de X(0) = x au dbut
    X = np.concatenate(([x], X))

    # Trac de la trajectoire
    plt.figure(figsize=(10, 6))
    plt.plot(t, X)
    plt.title('Trajectoire du mouvement brownien avec drive et variance')
    plt.xlabel('Temps t')
    plt.ylabel('X(t)')
    plt.grid(True)
    plt.show()

# Illustration
bm_drifted(T=1, mu=2, sigma=20, x=1)
```

```python
import numpy as np


a = 3
b = 5
n = 5000
dt = 0.01

# Simuler des trajectoires de mouvement brownien
L = [] # liste pour stocker les temps et les valeurs du mouvement brownien o -a ou b sont atteints

# boucle for pour crer n marches alatoires
for k in range(n):
    # initialisation de Bt et t  0
    Bt = 0
    t = 0
    # tant que le mouvement brownien n'atteint pas -a ou b, on continue le dplacement
    while True:
        Bt += np.random.normal(0, np.sqrt(dt))
        t += dt
        if Bt <= -a or Bt >= b:
            break
    if Bt >= b: # on stocke uniquement les valeurs de t et Bt lorsque Bt >= b dans la liste L
      L.append((t,Bt))

# Calcul de la probabilit d'atteindre b avant -a
taille = len(L) # on compte le nombre d'lements lorsque Bt >= b
P_Tb = taille / n # on divise par le nombre de simulation pour avoir le ratio exprimental
P_Tb_thorique = a / (a + b) # P( T = Tb ) = a / (a+b)

# Afficher les rsultats
print(f"P(T = T_b) exprimental = {P_Tb}")
print(f" P(T = T_b) thorique = {P_Tb_thorique}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

def interpolate_and_normalize_trajectory(x, N):
    interpolated = []
    for t in np.linspace(0, 1, N):
        nt = t * N
        k = int(np.floor(nt))
        fraction = nt - k
        if k == 0:
            interpolated.append(0)
        elif k < N:
            value = (1 / np.sqrt(N)) * (np.sum(x[:k]) + fraction * x[k])
            interpolated.append(value)
        else:
            value = (1 / np.sqrt(N)) * np.sum(x)
            interpolated.append(value)
    return np.array(interpolated)

N = 1000  # Nombre de pas de la marche alatoire

# Variables alatoires i.i.d. centres et rduites
X = np.random.normal(0, 1, N)

# Interpoler et normaliser la trajectoire
trajectory = interpolate_and_normalize_trajectory(X, N)

# Tracer la trajectoire
plt.figure(figsize=(12, 8))
plt.plot(np.linspace(0, 1, N), trajectory, label='Trajectoire interpole et normalise')

plt.title('Trajectoire normalise et interpole de la marche alatoire pour N = {}'.format(N))
plt.xlabel('Temps (t)')
plt.ylabel('Xn(t)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def interpolate_and_normalize_trajectory(x, N):
    interpolated = []
    for t in np.linspace(0, 1, N):
        nt = t * N
        k = int(np.floor(nt))
        fraction = nt - k
        if k == 0:
            interpolated.append(0)
        elif k < N:
            value = (1 / np.sqrt(N)) * (np.sum(x[:k]) + fraction * x[k])
            interpolated.append(value)
        else:
            value = (1 / np.sqrt(N)) * np.sum(x)
            interpolated.append(value)
    return np.array(interpolated)

N = 10000  # Nombre de pas de la marche alatoire (suffisamment grand pour illustrer la convergence)
num_trajectories = 10  # Nombre de trajectoires  tracer

# Gnration des variables alatoires i.i.d. centres et rduites
X = np.random.normal(0, 1, (num_trajectories, N))

# Interpoler et normaliser les trajectoires
trajectories = []
for i in range(num_trajectories):
    trajectory = interpolate_and_normalize_trajectory(X[i], N)
    trajectories.append(trajectory)

# Convertir en tableau numpy pour faciliter les calculs
trajectories = np.array(trajectories)

# Calculer la moyenne des trajectoires
mean_trajectory = np.mean(trajectories, axis=0)

# Tracer les trajectoires individuelles et la moyenne
plt.figure(figsize=(12, 8))
for i in range(num_trajectories):
    plt.plot(np.linspace(0, 1, N), trajectories[i], alpha=0.5)

plt.plot(np.linspace(0, 1, N), mean_trajectory, label='Moyenne des trajectoires', color='black', linewidth=2)

plt.title('Trajectoires normalises et interpoles de la marche alatoire (illustration du Thorme de Donsker)')
plt.xlabel('Temps (t)')
plt.ylabel('Xn(t)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Fonction pour gnrer un mouvement brownien standard
def mb(T, N):
    dt = T / N
    dW = np.sqrt(dt) * np.random.randn(N)
    W = np.cumsum(dW)
    W = np.concatenate(([0], W))  # Initialiser  W(0) = 0
    return W

# Fonction pour tracer des trajectoires du mouvement brownien gomtrique
def plot_mb_geometric(T, mu, sigma, N=1000, num_trajectories=5):
    t = np.linspace(0, T, N+1)

    plt.figure(figsize=(12, 8))

    for _ in range(num_trajectories):
        Bt = mb(T, N)
        S0 = 1 #np.random.uniform(0, 1)
        St = S0 * np.exp(mu * t - 0.5 * sigma**2 * t + sigma * Bt)
        plt.plot(t, St)

    plt.title('Trajectoires du mouvement brownien gomtrique')
    plt.xlabel('Temps (t)')
    plt.ylabel('S_t')
    plt.grid(True)
    plt.show()

# Paramtres
T = 50     # Borne suprieure de l'intervalle de temps
mu = 0.2      # Drift
sigma = 0.2    # Coefficient de diffusion
N = 100000       # Nombre de points de discrtisation
num_trajectories = 5  # Nombre de trajectoires  tracer

# Plot les trajectoires
plot_mb_geometric(T, mu, sigma, N, num_trajectories)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def pontbrownien(n, npas): # npas, nombre de pas et n, nombre de simulations donc de trajectoires
    dt = 1 / npas # calcul du pas en temps
    t = np.linspace(0, 1, npas + 1) # subdivision de l'intervalle [0,1]

    # Simuler n trajectoires de mouvement brownien
    B = np.random.normal(0, np.sqrt(dt), (n, npas)) # mouvement brownien classique
    B = np.cumsum(B, axis=1)                        # somme cumule pour obtenir les trajectoires
    B = np.hstack((np.zeros((n, 1)), B))            # On ajoute la condition B  t=0 vaut 0 pour chaque trajectoire

    # On cre le pont brownien en fixant B  t=1 ,  la valeur 0
    pont = B - t * B[:, -1][:, np.newaxis] # Comme Bt est un mouvement brownien, alors Bt - tB1 est un pont brownien d'aprs le cours
    # on rajoute [:, np.newaxis] par soucis de dimension

    return t, pont # on retourne la liste des temps et le pont brownien correspondant

# Paramtres de test
npas = 1000
n = 3

t, pont = pontbrownien(n, npas)

# Tracer les trajectoires
for i in range(n):
    plt.plot(t, pont[i])
plt.plot(t,np.zeros(npas+1), color='g')
plt.title('Trajectoires de pont brownien')
plt.xlabel('Temps')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Fonction pour gnrer un pont brownien sur [0, 1]
def brownian_bridge(T, N):
    dt = T / N
    t = np.linspace(0, T, N+1)
    W = np.random.normal(0, np.sqrt(dt), N)
    W = np.cumsum(W)
    W = np.concatenate(([0], W))
    W = W - t / T * W[-1]  # Adjust to make it a bridge
    return W

# Fonction pour gnrer une excursion brownienne en utilisant la transforme de Verwaat
def brownian_excursion(T, N):
    W = brownian_bridge(T, N)
    t = np.linspace(0, T, N+1)
    
    # Trouver l'argument du minimum de W
    Tmin = np.argmin(W)
    
    # Gnrer l'excursion brownienne Z_t
    Z = np.zeros_like(W)
    for i in range(N+1):
        if i <= Tmin:
            Z[i] = W[Tmin + i] - W[Tmin]
        else:
            Z[i] = W[Tmin + i - (N+1)] - W[Tmin]
    
    return t, Z

# Paramtres
T = 1.0        # Borne suprieure de l'intervalle de temps
N = 1000       # Nombre de points de discrtisation
num_trajectories = 5  # Nombre de trajectoires  tracer

# Tracer les trajectoires de l'excursion brownienne
plt.figure(figsize=(12, 8))
for _ in range(num_trajectories):
    t, Z = brownian_excursion(T, N)
    plt.plot(t, Z)

plt.title('Trajectoires de l\'excursion brownienne')
plt.xlabel('Temps (t)')
plt.ylabel('Z_t')
plt.grid(True)
plt.show()
```
